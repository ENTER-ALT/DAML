## Beschrijving

Beroerte blijft een van de grootste uitdagingen voor de volksgezondheid wereldwijd, vaak het gevolg van een combinatie van medische, leefstijl- en omgevingsfactoren. Het begrijpen van deze elementen is essentieel voor het ontwikkelen van effectieve preventiestrategieën en het verbeteren van gezondheidsresultaten.

Een beroerte ontstaat wanneer de bloedtoevoer naar een deel van de hersenen wordt onderbroken. Dit kan komen door een verstopping (zoals een bloedprop) of een bloeding in de hersenen. Doordat de hersenen bloed nodig hebben voor zuurstof en voedingsstoffen, kan een beroerte leiden tot hersenschade, invaliditeit of zelfs de dood.

Deze dataset onderzoekt de relatie tussen leefstijlfactoren en het risico op een beroerte en biedt een uitgebreid overzicht van individuele gezondheidsprofielen. Het bevat informatie over leeftijd, geslacht, hypertensie, hartziekten en familiegeschiedenis, naast leefstijlstatistieken zoals fysieke activiteit, eetgewoonten, roken, alcoholconsumptie en stressniveaus. Aanvullende details over BMI, glucoseniveaus, slaappatronen, burgerlijke staat, werktype en sociaaleconomische factoren bieden verdere context aan de analyse.

Elk record vertegenwoordigt de unieke gezondheidsreis van een individu, waarbij belangrijke variabelen worden vastgelegd die van invloed kunnen zijn op het risico op een beroerte. Met inzichten in risicoscores en -voorkomens van beroertes bieden deze gegevens een waardevolle kans om patronen te identificeren, risicofactoren te ontdekken en praktische interventies voor te stellen.

Jouw missie is om deze factoren te analyseren, belangrijke verbanden te benadrukken en bruikbare aanbevelingen te doen om een ​​gezondere levensstijl te ondersteunen en de prevalentie van beroertes te verminderen.

1. **ID**: anonieme identificatie
1. **Age**: leeftijd in jaren
1. **Gender**: man/vrouw
1. **Hypertension**: ja/nee
1. **Heart Disease**: hartziekte, ja/nee
1. **Ever Married**: ooit getrouwd, ja/nee
1. **Work Type**: werk categorieën: privé, overheid, zelfstandig, kinderen, nooit gewerkt
1. **Residence Type**: woning categorieën: stedelijk of landelijk
1. **Average Glucose Level**: bloedglucoseconcentratie in mg/dL
1. **BMI**: numerieke waarde die de verhouding weergeeft: gewicht/lengte*lengte
1. **Smoking Status**: rookstatus, categorieën: nooit gerookt, vroeger gerookt, rookt of onbekend
1. **Physical Activity**: Lichamelijke activiteit, categorieën: sedentair (geen beweging), licht, matig of actief
1. **Dietary Habits**: dieetgewoonten, categorieën: natriumarm dieet, bewerkt/geraffineerd dieet, vetrijk/cholesterolrijk dieet, plantaardig dieet, mediterraan dieet
1. **Alcohol Consumption**: ja/nee
1. **Chronic Stress**: ja/nee
1. **Sleep Hours**: gemiddeld aantal uren slaap per dag
1. **Family History of Stroke**: familiegeschiedenis van beroerte, ja/nee
1. **Education Level**: onderwijsniveau, categorieën: geen onderwijs, primair (tot 12 jaar), secundair (13-18 jaar oud) of tertiair (post-secundair)
1. **Income Level**: inkomensniveau, categorieën: laag, gemiddeld of hoog
1. **Region**: deel van het land waar ze wonen - categorieën: noord, zuid, oost, west
1. **Stroke Occurrence**: beroerte-voorkomen, ja/nee
